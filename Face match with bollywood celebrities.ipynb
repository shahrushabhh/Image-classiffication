## KNN ALGORITHM

```python
def dist(a1,a2):
    return np.sum((a1-a2)**2)**.5
```

```python
def KNN(X,Y,test_point,k=5):
    tt = cv2.imread('C:/Users/Rushabh Shah/Downloads\Boolywood celebrity face/Train/Images/' + X[0])
    t2 = cv2.resize(tt,(200,200))
    m=t2.shape[0]
    vals = []
    for i in range(200):
        con = cv2.imread('C:/Users/Rushabh Shah/Downloads\Boolywood celebrity face/Train/Images/' + X[i])
        con1 = cv2.cvtColor(con, cv2.COLOR_BGR2RGB)
        con2 = cv2.resize(con1,(200,200))
        d = dist(con2,test_point)
        vals.append((d,Y[i]))
    
    vals = sorted(vals)
    vals = vals[:k]
    
    return vals
    
    #vals = np.array(vals[:k])
    #b= np.unique(vals[:,1],return_counts=True)
    #idx = np.argmax(b[i])
    #pred = b[0][idx]
    
    #return str(pred)
```
